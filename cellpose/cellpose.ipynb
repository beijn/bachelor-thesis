```python
# see https://cellpose.readthedocs.io/en/latest/notebook.html
# and https://cellpose.readthedocs.io/en/latest/outputs.html#plotting-functions

import numpy as np
import matplotlib.pyplot as plt
from cellpose import models
from cellpose.io import imread
from cellpose import plot, utils


model = models.Cellpose(model_type='cyto', gpu=True, net_avg=True)

imgs = [imread(f) for f in ['../data/first/PC-10x-wide.jpg']]
nimg = len(imgs)

channels = [[0,0]]

masks, flows, styles, diams = model.eval(imgs, diameter=None, channels=channels)  # automatic diameter ~70 was best
for idx in range(nimg):
  maski = masks[idx]
  flowi = flows[idx][0]

  fig = plt.figure(figsize=(12,5))
  plot.show_segmentation(fig, imgs[idx], maski, flowi, channels=channels[idx])
  plt.tight_layout()
  plt.show()
```
